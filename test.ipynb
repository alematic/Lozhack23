```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://transportal.compbio.ucsf.edu/index/'
res = requests.get(url)
# print(res.text) # Uncomment to see the HTML


soup = BeautifulSoup(res.text, 'html.parser')

# Find all <a> tags and filter for hrefs that meet criteria
transporter_entries = soup.find_all('a', href=True)
#what is the type of transporter_entries
type(transporter_entries)

#filter from tranpsorter_entries all the entries that have /transporters/ inside
transporter_entries = [entry for entry in transporter_entries if '/transporters/' in entry.get('href')]

#filter out also the mouse and rat entries
transporter_entries = [entry for entry in transporter_entries if not '/transporters/mouse' in entry.get('href')]
transporter_entries = [entry for entry in transporter_entries if not '/transporters/rat' in entry.get('href')]
transporters = pd.DataFrame(transporter_entries, columns=['name'])
```

```python
transporters
```

```python
sqce=[]
for name in transporters['name']:
    response = requests.get('https://rest.uniprot.org/uniprotkb/search?query='+name+'+AND+organism_id:9606&includeisoform=false')
    response = response.json()['results']
    sqce.append(response[0]['sequence']['value'])
transporters
```
